### Basisgegevens verzamelen

```python
def parse_car_details_regex(text):
    patterns = {
        'Carrosserie': r'Carrosserietype(\w+?)(?=Categorie|Aandrijving|Aantal|Deuren|$)',
        'Categorie': r'Categorie(\w+?)(?=Carrosserietype|Aandrijving|Aantal|Deuren|$)',
        'Aandrijving': r'Aandrijving(\w+?)(?=Carrosserietype|Categorie|Aantal|Deuren|$)',
        'Aantal_stoelen': r'Aantal\s*stoelen(\d+)',
        'Deuren': r'Deuren(\d+)'
    }
    
    results = {}
    
    for key, pattern in patterns.items():
        match = re.search(pattern, text)
        results[key] = match.group(1) if match else 'Not found'
    
    return results

text = data_items[0].text.strip()
results_regex = parse_car_details_regex(text)

for key, value in results_regex.items():
    print(f"{key.replace('_', ' ')}: {value}")
```

### Voertuiggeschidenis (problemen met vinden onderhoudshistoriek)

```python
def parse_car_details_regex(text):
    patterns = {
        'Kilometerstand': r'Kilometerstand([\d.]+)\s*km',
        'Bouwjaar': r'Bouwjaar(\d{2}/\d{4})',
        'Vorige_eigenaren': r'Vorige\s*eigenaren(\d+)',
        'Onderhoudshistoriek': r'Onderhoud(?:s)?historiek(Ja|Nee)',  # Made more flexible
        'Niet_rokers_auto': r'Niet-rokers\s*auto(Ja|Nee)'
    }
    
    results = {}
    
    for key, pattern in patterns.items():
        match = re.search(pattern, text)
        results[key] = match.group(1) if match else 'Not found'
        
        if key == 'Kilometerstand' and match:
            results[key] = f"{results[key]} km"
    
    return results

text = data_items[1].text.strip()
results_regex = parse_car_details_regex(text)

for key, value in results_regex.items():
    print(f"{key.replace('_', ' ')}: {value}")
```

### Technische gegevens

```python
def parse_car_details_regex(text):
    patterns = {
        'Vermogen': r'Vermogen(\d+\s*kW\s*\(\d+\s*PK\))',
        'Transmissie': r'Transmissie(\w+?)(?=Vermogen|Cilinderinhoud|Versnellingen|Cilinders|Gewicht|$)',
        'Cilinderinhoud': r'Cilinderinhoud([\d.]+\s*cm³)',
        'Versnellingen': r'Versnellingen(\d+)',
        'Cilinders': r'Cilinders(\d+)',
        'Gewicht': r'Gewicht([\d.]+\s*kg)'
    }
    
    results = {}
    
    for key, pattern in patterns.items():
        match = re.search(pattern, text)
        results[key] = match.group(1) if match else 'Not found'
        
        if match:
            results[key] = ' '.join(results[key].split())
    
    return results

text = data_items[2].text.strip()
results_regex = parse_car_details_regex(text)

for key, value in results_regex.items():
    print(f"{key}: {value}")
```

### Energieverbruik

```python
def parse_car_details_regex(text):
    patterns = {
        'Emissieklasse': r'Emissieklasse(Euro\s*\d+\w*?)(?=Brandstof|Brandstofverbruik|CO2-emissie|$)',
        'Brandstof': r'Brandstof(\w+?)(?=Emissieklasse|Brandstofverbruik|CO2-emissie|$)',
        'Brandstofverbruik': r'Brandstofverbruik([\d,]+\s*l/100\s*km\s*\(comb\.\))',
        'CO2_emissie': r'CO2-emissie([\d,]+\s*g/km\s*\(comb\.\))'
    }
    
    ordered_keys = ['Emissieklasse', 'Brandstof', 'Brandstofverbruik', 'CO2_emissie']
    results = {}
    
    for key in ordered_keys:
        pattern = patterns[key]
        match = re.search(pattern, text)
        results[key] = match.group(1) if match else 'Not found'
        
        if match:
            results[key] = ' '.join(results[key].split())
    
    return results

text = data_items[3].text.strip()
results_regex = parse_car_details_regex(text)

for key, value in results_regex.items():
    print(f"{key.replace('_', ' ')}: {value}")
```

### Uitrusting

```python
def parse_car_details_regex(text):
    patterns = {
        'Comfort_en_gemak': r'Comfort en gemak([\w\s\-/]+?)(?=Entertainment en Media|Veiligheid|Extra|$)',
        'Entertainment_en_Media': r'Entertainment en Media([\w\s\-/]+?)(?=Comfort en gemak|Veiligheid|Extra|$)',
        'Veiligheid': r'Veiligheid([\w\s\-/]+?)(?=Comfort en gemak|Entertainment en Media|Extra|$)',
        'Extra': r'Extra([\w\s\-/]+?)(?=Comfort en gemak|Entertainment en Media|Veiligheid|$)'
    }

    ordered_keys = ['Comfort_en_gemak', 'Entertainment_en_Media', 'Veiligheid', 'Extra']
    results = {}

    for key in ordered_keys:
        pattern = patterns[key]
        match = re.search(pattern, text)
        results[key] = match.group(1).strip() if match else 'Not found'

    return results

text = data_items[4].text.strip()

results_regex = parse_car_details_regex(text)

for key, value in results_regex.items():
    print(f"{key.replace('_', ' ')}: {value}")
```

### Kleur en bekleding

```python
def parse_car_details_regex(text):
    patterns = {
        'Koetswerkkleur': r'Koetswerkkleur(\w+)(?=Soort lak|Kleur interieur|Interieur|$)',
        'Soort_lak': r'Soort lak(\w+)(?=Koetswerkkleur|Kleur interieur|Interieur|$)',
        'Kleur_interieur': r'Kleur interieur(\w+?)(?=Interieur|Koetswerkkleur|Soort lak|$)',
        'Interieur': r'Interieur((?:\w+\s*)+)(?=Koetswerkkleur|Soort lak|Kleur interieur|$)'
    }

    ordered_keys = ['Koetswerkkleur', 'Soort_lak', 'Kleur_interieur', 'Interieur']
    results = {}

    for key in ordered_keys:
        pattern = patterns[key]
        match = re.search(pattern, text)
        results[key] = match.group(1) if match else 'Not found'
        
        if match:
            results[key] = results[key].strip()

    return results

text = data_items[5].text.strip()
results_regex = parse_car_details_regex(text)

for key, value in results_regex.items():
    print(f"{key.replace('_', ' ')}: {value}")
```
